```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_list_from_file(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
        return eval(content)

def plot_2d_array(array):
    plt.imshow(array, cmap='viridis', aspect='auto')
    plt.colorbar()
    plt.show()
```

```python
data = read_list_from_file('received_data.txt')
data = data[:261120]
```

```python
data_2d = np.flipud(np.array(data).reshape(480, 544))
print(data_2d.shape)
```

```python
plot_2d_array(data_2d)
```

```python
data_2d[data_2d<0.35] = 10
plot_2d_array(data_2d)
```
